```python
import pandas as pd

df = pd.read_csv("03_LibrarySystembook.csv")

df = df.dropna(how='any')

df['Books'] = df['Books'].str.title()

df['Book checkout'] = df['Book checkout'].str.replace('"', '', regex=False)

df['Days allowed to borrow'] = df['Days allowed to borrow'].replace('2 weeks',14)

df['Books'] = df['Books'].str.strip()

df['Id'] = pd.to_numeric(df['Id'], errors='coerce').astype('Int64')

df['Customer ID'] = pd.to_numeric(df['Customer ID'], errors='coerce').astype('Int64')

df = df.rename(columns={

    'Id': 'BOOK_ID',
    'Books': 'BOOK_NAME',
    'Book checkout': 'CHECKOUT_DATE',
    'Book Returned': 'RETURN_DATE',
    'Days allowed to borrow': 'BORROW_DAYS',
    'Customer ID': 'CUSTOMER_ID',

})

df.to_csv("LibraryBooksCleaned.csv", index=False)

print(df)
```
