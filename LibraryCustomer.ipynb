```python
import pandas as pd

df = pd.read_csv("03_Library SystemCustomers.csv")

df = df.dropna(how='any')

df['Customer Name'] = df['Customer Name'].str.title()

df['Customer Name'] = df['Customer Name'].str.strip()

df['Customer ID'] = pd.to_numeric(df['Customer ID'], errors='coerce').astype('Int64')

df = df.rename(columns={

    'Customer ID': 'CUSTOMER_ID',
    'Customer Name': 'CUSTOMER_NAME'

})

df.to_csv("LibraryCustomerCleaned.csv", index=False)

print(df)
```
